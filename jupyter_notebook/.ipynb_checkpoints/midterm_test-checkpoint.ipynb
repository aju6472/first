```python
import matplotlib.pyplot as plt #12
from matplotlib import rc
import random

rc('font', family='Malgun Gothic')

def make_random_list(num):
    result = []
    for i in range(num):
        result.append(random.randrange(1,7))
    return result

count = 10
y = make_random_list(count)
x = list(range(1,11))

plt.title('주사위 10회 던지기')
plt.bar(x,y)
plt.yticks([1,2,3,4,5,6])
plt.xlabel('던진 횟수')
plt.ylabel('주사위 눈')

plt.show()
```

```python
import csv #15

f = open('cafe_2year.csv', 'r', encoding = 'utf-8')
lines = csv.reader(f)

next(lines)

z=1
xdata = []
ydata = []

for line in lines :
    if z % 3 == 0 :
        xdata.append(line[0])
        ydata.append(line[1])
    z += 1

import matplotlib.pyplot as plt 
from matplotlib import rc

rc('font', family='Malgun Gothic')

font1 = {'size':18, 'color':'green'}

plt.bar(xdata,ydata)
plt.title('전국의 월별 까페 수', fontdict = font1)
plt.xlabel('년월')
plt.ylabel('카페수')

plt.show()
f.close()
```

```python
import csv #Q2,3

file1 = open('doctor_2019.csv', 'r', encoding = 'utf-8') 
file2 = open('doctor2.csv','w',encoding = 'utf-8',newline='')
lines = csv.reader(file1)
wr = csv.writer(file2)

header=next(lines)

a = (['서울',0,0,0,0],['부산',0,0,0,0],['대구',0,0,0,0],['인천',0,0,0,0],['대전',0,0,0,0],['광주',0,0,0,0],['울산',0,0,0,0])
for line in lines:
    if line[0] == '서울' :
        a[0][1] += int(line[2])
        a[0][2] += int(line[3])
        a[0][3] += int(line[4])
        a[0][4] += int(line[5])
    if line[0] == '부산' :
        a[1][1] += int(line[2])
        a[1][2] += int(line[3])
        a[1][3] += int(line[4])
        a[1][4] += int(line[5])
    if line[0] == '대구' :
        a[2][1] += int(line[2])
        a[2][2] += int(line[3])
        a[2][3] += int(line[4])
        a[2][4] += int(line[5])
    if line[0] == '인천' :
        a[3][1] += int(line[2])
        a[3][2] += int(line[3])
        a[3][3] += int(line[4])
        a[3][4] += int(line[5])
    if line[0] == '대전' :
        a[4][1] += int(line[2])
        a[4][2] += int(line[3])
        a[4][3] += int(line[4])
        a[4][4] += int(line[5])
    if line[0] == '광주' :
        a[5][1] += int(line[2])
        a[5][2] += int(line[3])
        a[5][3] += int(line[4])
        a[5][4] += int(line[5])
    if line[0] == '울산' :
        a[6][1] += int(line[2])
        a[6][2] += int(line[3])
        a[6][3] += int(line[4])
        a[6][4] += int(line[5])
        
print(header[0], header[2], header[3], header[4], header[5])
for x in a:
    print('%s %8d %8d %8d %8d' %(x[0],x[1],x[2],x[3],x[4]))
    
wr.writerow(['지역','일반의','인턴','레지던트','전문의'])
for i in range(7):
    wr.writerow([a[i][0],a[i][1],a[i][2],a[i][3],a[i][4]])

file1.close()
file2.close()
```

```python
#17
f = open('doctor2.csv', 'r', encoding = 'utf-8')
lines = csv.reader(f)

import matplotlib.pyplot as plt 
from matplotlib import rc

rc('font', family='Malgun Gothic')
font1 = {'size':18, 'color':'green'}

xdata = [] 
ydata1 = [] #일반의
ydata2 = [] #인턴
ydata3 = [] #레지던트
ydata4 = [] #전문의

next(lines)
for line in lines:
    xdata.append(line[0])
    ydata1.append(int(line[1]))
    ydata2.append(int(line[2]))
    ydata3.append(int(line[3]))
    ydata4.append(int(line[4]))


plt.title('서울 및 광역시 종합 병원 전문의 수', fontdict = font1)
plt.plot(xdata, ydata1, label = '일반의', color='red', linestyle='-', marker='o')
plt.plot(xdata, ydata2, label = '인턴', color='yellow', linestyle='--', marker='s')
plt.plot(xdata, ydata3, label = '레지던트', color='blue', linestyle=':', marker='x')
plt.plot(xdata, ydata4, label = '전문의', color='magenta', linestyle='-.', marker='d')
plt.xlabel('시')
plt.ylabel('의사 수')
plt.show()

f.close()

```

```python
#18
f = open('doctor2.csv', 'r', encoding = 'utf-8')
lines = csv.reader(f)

import matplotlib.pyplot as plt 
from matplotlib import rc

rc('font', family='Malgun Gothic')
font1 = {'size':18, 'color':'green'}

doctors = ['일반의','인턴','레지던트','전문의']
numbers = []

header = next(lines)

seoul = next(lines)

total = 0
for i in range(4) :
    total += int(seoul[i+1])
    
for i in range(4) :
    ratio = int(seoul[i+1])/total;
    numbers.append(ratio)

plt.pie(numbers, explode=(0, 0.1, 0, 0), labels=doctors, autopct='%.1f%%', shadow=True, startangle=90)
plt.title('서울 종합병원 의사 유형별 분포', fontdict=font1)

plt.show()
        
f.close()
```

```python
#19
import csv

f = open('month_temp.csv', 'r', encoding = 'utf-8')
lines = csv.reader(f)

import matplotlib.pyplot as plt 
from matplotlib import rc

rc('font', family ='Malgun Gothic')
font1 = {'size':18, 'color':'green'}

xdata = []
ydata = []

next(lines)

count = 1
for line in lines :
    if count % 3 == 0 :
        xdata.append(line[1][8:])
        ydata.append(line[2])
    count += 1
    
        
plt.plot(xdata,ydata)
plt.title('5일 간의 평균 기온', fontdict = font1)
plt.xlabel('일')
plt.ylabel('평균 기온')

plt.show
f.close()
```

```python
#20
import csv

f = open('jeju_2019.csv', 'r', encoding='utf-8')
lines = csv.reader(f)


rc('font', family ='Malgun Gothic')
font1 = {'size':18, 'color':'green'}

header = next(lines)

total_rain = [0, 0, 0, 0]

for line in lines:
    if not line[5] :             
        line[5] = 0
        
    if line[1] == '제주' :    # '제주' 지역
        total_rain[0] += float(line[5])
        
    if line[1] == '고산' :    # '고산' 지역
        total_rain[1] += float(line[5])
        
    if line[1] == '성산' :    # '성산' 지역
        total_rain[2] += float(line[5])
        
    if line[1] == '서귀포' :    # '서귀포' 지역
        total_rain[3] += float(line[5])
        
xdata = ['제주','고산','성상','서귀포']
ydata = [total_rain[0],total_rain[1],total_rain[2],total_rain[3]]

plt.plot(xdata,ydata)
plt.title('지역 별 연 강수량', fontdict = font1)
plt.xlabel('일')
plt.ylabel('연 강수량')

plt.show()
f.close()
```
